```python
import pandas as pd
import matplotlib.pyplot as plt

datos = pd.read_csv("School-Age-Digital-Connectivity - Primary.csv")

df = datos[["Countries and areas","Region","Income Group","Total"]]

df = datos.rename(columns = {
    "Countries and areas":"PAÍSES",
    "Region":"REGIONES",
    "Income Group":"GRUPOS",
    "Total":"TOTAL"
})
#print(df)
# Contar el número de veces que aparece la región "SSA"
region_countSSA = (df['REGIONES'] == 'LAC').sum()
# Contar el número de veces que aparece la región "LAC"
region_countLAC = (df['REGIONES'] == 'LAC').sum()
# Contar el número de veces que aparece la región "ECA"
region_countECA = (df['REGIONES'] == 'ECA').sum()
# Contar el número de veces que aparece la región "SA"
region_countSA = (df['REGIONES'] == 'SA').sum()

# Contar el número de veces que aparece la región "EAP"
region_countEAP = (df['REGIONES'] == 'EAP').sum()
total_rows = df['REGIONES'].count()

# Datos para el gráfico de pastel
labels = ['SSA', 'LAC', 'ECA', 'SA', 'EAP']
sizes = [region_countSSA/total_rows, region_countLAC/total_rows, region_countECA/total_rows, region_countSA/total_rows, region_countEAP/total_rows]
colors = ['red', 'blue', 'green', 'yellow', 'orange']

# Crear el gráfico de pastel
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)

# Añadir un título
plt.title('Distribución de regiones')

# Añadir una leyenda con descripciones en una posición ajustada
descriptions = ['(SSA) África Subsahariana', '(LAC) América Latina y el Caribe', '(ECA) Europa y Asia Central', '(SA) Asia Meridional', '(EAP) Asia Oriental y el Pacífico']
plt.legend(descriptions, loc='center left', bbox_to_anchor=(1, 0.5))

# Mostrar el gráfico
plt.show()


```
